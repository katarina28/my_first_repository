Now let's merge the data frames v_product_list_available_simple_in_stock_basic and stock_days_data_in_stock_basic.
Since I only need data for simple and available products, I will need to have displayed each row from the v_product_list_available_simple_basic table and matching number of rows of the stock_days_data_in_stock_basic table

Final table with product IDs and average quantity of the product in stock as per date

```python
Series = available_simple_products_in_stock_simple.groupby(['entity_id_hash', 'period'])['qty'].mean()
Final = Series.to_frame('avg quantity')
Final.head(100)
```
